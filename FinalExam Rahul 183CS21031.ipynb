```python
#Impoting Packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
#Loading Dataset
bc=load_breast_cancer()
n_data=pd.DataFrame(data=bc.data,columns=bc.feature_names)
n_data
```

```python
y=pd.Categorical.from_codes(bc.target,bc.target_names)mm
y
```

```python
#then training the model
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(n_data,y,test_size=0.2,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#fitting the model using logistic regression
from sklearn.linear_model import LogisticRegression
```

```python

from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model=LogisticRegression(max_iter=10000)
```

```python
model.fit(x_train,np.ravel(y_train))
```

```python
y_pred=model.predict(x_train)
```

```python
#importing metrics for evaluate the model
from sklearn.metrics import confusion_matrix,classification_report
```

```python
conf_m=confusion_matrix(y_train,y_pred)
conf_m
```

```python
report=classification_report(y_train,y_pred)
print(report)
```

```python
#checking the testing set 
y_test_pred=model.predict(x_test)
```

```python
conf_m=confusion_matrix(y_test,y_test_pred)
conf_m
```

```python
report=classification_report(y_test,y_test_pred)
print(report)
```
